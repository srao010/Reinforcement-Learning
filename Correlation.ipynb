```python
import numpy as np
import random
```

```python
np.random.seed(1) #function to select random values
```

```python
#1000 random integers between 0 and 50
x = np.random.randint(0,50,1000) #100 values between 0 to 50
```

```python
x
```

```python
#Positive correlation with some noise
y = x + np.random.normal(0,10,1000)
y
```

```python
np.corrcoef(x,y) #positive correlation
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python

y = x - np.random.normal(0,10,1000)
y
```

```python
np.corrcoef(x,y) #negative correlation
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
#Negative correlation
#1000 random integers between 0 and 50
x = np.random.randint(0,50,1000) #1000 values between 0 to 50
```

```python
y = 100 - x +np.random.normal(0,5,1000)
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)
np.corrcoef(x,y) # 0 correlation
```

```python
import pandas as pd
df = pd.DataFrame({'a': np.random.randint(0,50,1000)})
df['b'] = df['a'] + np.random.randint(0,10,1000) #+ve
df['c'] = 100- df['a'] + np.random.randint(0,5,1000) #-ve
df['d'] = np.random.randint(0,50,1000) #no corr

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize = (6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
